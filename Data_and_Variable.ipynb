# Module 1: Data and Variables


## 1. Introduction
Data analysis is essential for extracting meaningful insights from various fields of study. Understanding the types of data and variables is the foundation for effective analysis.

---

## 2. Definitions of Data and Variable

- **Data**: Data refers to raw facts and figures that can be processed to obtain meaningful information.
- **Variable**: A variable is a characteristic or attribute that can take on different values. Variables can change and are measured in data analysis.

---

## 3. Types of Data


**Data** can be devided in to various types based on structure, attribute, etc.
### i. Based on structure:

| **Data Type**       | **Definition**                                                | **Examples**                                     |
|---------------------|--------------------------------------------------------------|--------------------------------------------------|
| **Unstructured Data** | Data that lacks a predefined format or organization. It is often text-heavy and requires more complex processing to extract meaningful information. | Emails, social media posts, videos, images, audio files |
| **Structured Data**  | Data that is organized in a predefined manner, typically in tables with rows and columns. Easily searchable and analyzable using standard database tools. | SQL databases, Excel spreadsheets, CSV files    |
| **Semi-Structured Data** | Data that does not have a rigid structure but contains tags or markers to separate semantic elements. This allows for some level of organization while still being flexible. | JSON files, XML files, HTML documents            |

| **Data Type**       | **Definition**                                                | **Examples**                                     |
|---------------------|--------------------------------------------------------------|--------------------------------------------------|

| **Quantitative Data**| Numeric data that can be measured and expressed as a number. | Height, weight, temperature, age                 |
| **Nominal Data**    | A subtype of qualitative data that categorizes without a natural order. | Gender (male, female), eye color (blue, green)  |

| **Continuous Data** | A subtype of quantitative data that can take on an infinite number of values within a given range. | Height (e.g., 170.5 cm), time (e.g., 3.75 hours) |

## 7. Data Visualization

### Bar Chart Example



```python
import numpy as np
import matplotlib.pyplot as plt

# Provided data
data = np.array([11, 12, 45, 10, 15, 12, 12, 18, 16, 15, 25, 15, 18, 30, 18, 21, 32, 36, 21, 33, 21, 40, 21, 33, 21])

# Get unique values and their frequencies for ungrouped data
unique, counts = np.unique(data, return_counts=True)

# Create the subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# 1. Ungrouped Frequency Distribution
axs[0, 0].bar(unique, counts, color='blue', alpha=0.7)
axs[0, 0].set_title('Ungrouped Frequency Distribution')
axs[0, 0].set_xlabel('Age(in years)')
axs[0, 0].set_ylabel('Frequency')
axs[0, 0].grid(axis='y', linestyle='--', alpha=0.7)

# 2. Cumulative Frequency Distribution (Ungrouped)
cumulative_counts = np.cumsum(counts)
axs[0, 1].bar(unique, cumulative_counts, color='green', alpha=0.7)
axs[0, 1].set_title('Cumulative Frequency Distribution (Ungrouped)')
axs[0, 1].set_xlabel('Age(in years)')
axs[0, 1].set_ylabel('Cumulative Frequency')
axs[0, 1].grid(axis='y', linestyle='--', alpha=0.7)

# 3. Grouped Frequency Distribution
# Define bins for grouping
bins = [10, 15, 20, 25, 30, 35, 40, 45]
counts, bin_edges = np.histogram(data, bins=bins)
axs[1, 0].bar(bin_edges[:-1], counts, width=np.diff(bin_edges), color='orange', alpha=0.7, align='edge')
axs[1, 0].set_title('Grouped Frequency Distribution')
axs[1, 0].set_xlabel('Age Groups')
axs[1, 0].set_ylabel('Frequency')
axs[1, 0].grid(axis='y', linestyle='--', alpha=0.7)

# 4. Cumulative Frequency Distribution (Grouped)
cumulative_counts_grouped = np.cumsum(counts)
axs[1, 1].bar(bin_edges[:-1], cumulative_counts_grouped, width=np.diff(bin_edges), color='purple', alpha=0.7, align='edge')
axs[1, 1].set_title('Cumulative Frequency Distribution (Grouped)')
axs[1, 1].set_xlabel('Age Groups')
axs[1, 1].set_ylabel('Cumulative Frequency')
axs[1, 1].grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Provided data
data = np.array([11, 12, 45, 10, 15, 12, 12, 18, 16, 15, 25, 15, 18, 30, 18, 21, 32, 36, 21, 33, 21, 40, 21, 33, 21])

# Define bins for grouping
bins = [10, 15, 20, 25, 30, 35, 40, 45]
counts, bin_edges = np.histogram(data, bins=bins)

# Calculate midpoints of each bin
midpoints = 0.5 * (bin_edges[:-1] + bin_edges[1:])

# Create the frequency polygon
plt.figure(figsize=(8, 5))
plt.plot(midpoints, counts, marker='o', linestyle='-', color='blue', alpha=0.7)
plt.title('Frequency Polygon of Grouped Data')
plt.xlabel('Midpoints of Age Groups')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(midpoints)  # Set x-ticks to midpoints for clarity
plt.show()
```

```python
# Provided data
data = np.array([11, 12, 45, 10, 15, 12, 12, 18, 16, 15, 25, 15, 18, 30, 18, 21, 32, 36, 21, 33, 21, 40, 21, 33, 21])

# Get unique values and their frequencies
unique, counts = np.unique(data, return_counts=True)

# Pie Chart for Unique Values
plt.figure(figsize=(8, 8))
plt.pie(counts, labels=unique.astype(str), autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Pie Chart of Unique Values')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Define bins for grouping
bins = [10, 15, 20, 25, 30, 35, 40, 45]
counts_grouped, bin_edges = np.histogram(data, bins=bins)

# Pie Chart for Grouped Data
plt.figure(figsize=(8, 8))
plt.pie(counts_grouped, labels=[f"{bins[i]}-{bins[i+1]}" for i in range(len(bins)-1)], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Pie Chart of Grouped Age Ranges')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(range(len(data)), data, color='orange', alpha=0.6)
plt.title('Scatter Plot of Data')
plt.xlabel('Index')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

Here's a summary of the various data visualization techniques:

| Visualization Type         | Meaning                                         | Uses                                                      |
|----------------------------|------------------------------------------------|----------------------------------------------------------|
| **Bar Chart**              | Displays categorical data with rectangular bars. | Compare frequencies or counts of different categories.    |
| **Histogram**              | Shows the distribution of numerical data using bars. | Visualize the frequency distribution of continuous data.  |
| **Box Plot**               | Summarizes data using median, quartiles, and outliers. | Identify data spread, central tendency, and outliers.     |
| **Pie Chart**              | Circular chart divided into sectors representing proportions. | Show proportions of a whole (e.g., percentages of categories). |
| **Scatter Plot**           | Displays values for two variables as points on a Cartesian plane. | Identify relationships or correlations between two variables. |
| **Violin Plot**            | Combines box plot and density plot to show data distribution. | Visualize data distribution and its density.               |
| **Frequency Polygon**      | Line graph connecting midpoints of class intervals. | Show the distribution shape of grouped data.              |
| **Cumulative Frequency Plot** | Shows cumulative frequency up to each value or group. | Visualize how frequencies accumulate over data rand their applications.
